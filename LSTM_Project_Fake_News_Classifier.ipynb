```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
```

```python
len(messages)
```

```python
corpus = []
lem = WordNetLemmatizer()
for sentence in range(len(messages)):
    text = re.sub('[^a-zA-Z]', ' ', messages['title'][sentence])
    text = text.lower()
    text = text.split()    # splits the sentence into words
    text = [lem.lemmatize(word) for word in text if not word in set(stopwords.words('english'))]
    text = ' '.join(text)
    corpus.append(text)
    
```

```python
corpus
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
```

```python
voc_size = 5000
onehot_repr = [one_hot(sentence, voc_size) for sentence in corpus]
print(onehot_repr)
```

```python
sent_length = 20
equal_length_sentence = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length )
equal_length_sentence
```

```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features))
model.add(LSTM(100))  #an lstm layer with 100 neurons
model.add(Dense(1, activation = 'sigmoid'))    # an output layer with one neuron with sigmoid activation function because output is either 0 or 1
model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.build((None, sent_length))
model.summary()
```

```python
import numpy as np

X_final = np.array(equal_length_sentence)
Y_final = np.array(Y)
```

```python
print(X_final.shape, Y_final.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_final, Y_final, test_size=0.33, random_state=0)
```

```python
model.fit(x_train, y_train, batch_size=64, epochs= 20)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.where(y_pred>0.5 , 1,0) # becuase sigmoid predicts values from 0 to 1 we will convert them either to 0 or 1
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
score = accuracy_score(y_test, y_pred)
print(score)
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_test, y_pred))
```
